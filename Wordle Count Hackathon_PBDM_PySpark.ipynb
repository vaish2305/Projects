Task 0: Find the number of list of 10 words in the doc. 

```python
# data is moved to rdd "a" with lowering all the characters and removing special characters.
def lower_clean_str(x):
    char = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}—~“”’'
    lowercased_str = x.lower()
    for ch in char:
        lowercased_str = lowercased_str.replace(ch, '')
    return lowercased_str
# create Rdd (VAR1) from the file in Github
var1 = spark.sparkContext.textFile('/content/temp1/PLAW-117publ58 (1) (2).txt').map(lower_clean_str)
var1.top(10)
```

```python
# Tokenized all the lines 
words = var1.flatMap(lambda line: line.split())

# MapReduced to find frequency
wordCounts = words.map(lambda word: (word, 1))
wordCounts = wordCounts.reduceByKey(lambda x, y: x + y)
```

```python
# Take sample of random 10 words
random_words = wordCounts.takeSample(False, 10)

# Print random 10 words
print("Any 10 words")
print("--------------------------")
for word, count in random_words:
    print("{}: {}".format(word, count))
```

Task 1: Find 10 most frequent words in the bill.


```python
# Print top 10 most frequent words
topWords = wordCounts.takeOrdered(10, lambda x: -x[1])

print("Top 10 most frequent words")
print("--------------------------")
for word, count in topWords:
    print("{}: {}".format(word, count))
```

Task 2: Find 10 most frequent wordle word in the bill (note that wordle is a valid word that has five letters, no digits, such as grant, award, and funds.

```python
# Filter the wordCounts with only wordle word
wordleWords = wordCounts.filter(lambda x: x[0].isalpha() and len(x[0]) == 5 and not any(char.isdigit() for char in x[0]))

# Print top 10 most frequent wordle word
topWodleWords = wordleWords.takeOrdered(10, lambda x: -x[1])

print("Top 10 most frequent wordle word")
print("--------------------------------")
for word, count in topWodleWords:
    print("{}: {}".format(word, count))
```

Task 3: (TO DO): Find all wordles that start and end with same letter.

```python
# Filter if wordle words start and end with same letter
sameWordleWords = wordleWords.filter(lambda x: x[0][0] == x[0][-1]).collect()

# Sort sameWordleWords in descending order based on their counts
sorted_sameWordleWords = sorted(sameWordleWords, key=lambda x: x[1], reverse=True)

# Print  all wordles that start and end with same letter
print("All wordle words start and end with same letter")
print("-----------------------------------------------")
for word, count in sorted_sameWordleWords:
    print("{}: {}".format(word, count))
```

Task 4: (TO DO): Find all wordles where all 5 letters are distinct.

```python
# Filter if all wordles where all 5 letters are distinct
distinctWordleWords = wordleWords.filter(lambda x: len(set(x[0])) == 5).collect()

# Sort distinctWordleWords in descending order based on their counts
sorted_distinctWordleWords = sorted(distinctWordleWords, key=lambda x: x[1], reverse=True)

# Print all wordles where all 5 letters are distinct
print("All wordle words where all 5 letters are distinct")
print("-------------------------------------------------")
for word, count in sorted_distinctWordleWords:
    print("{}: {}".format(word, count))
```
