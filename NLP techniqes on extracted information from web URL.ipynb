# 2Q. Extract information from the web URL using BeautifulSoup library and display the content.

```python
# Import requests and BeautifulSoup library to extract the info from html page.
import requests 
from bs4 import BeautifulSoup 

# Get the content from the website.
r = requests.get('https://www.storynory.com/fox-and-the-crow/')

# Read the content in html parser.
soup = BeautifulSoup(r.content, 'html.parser') 

par = ""
# Read all the <p> tags and concatenate all the <p> tags and form a story.
for data in soup.find_all("p"): 
    par+=data.get_text() 

# Display the <p> tag content.
print(par) 
```

### 1. Tokenization: Convert paragraph to sentences and sentences to words.
    
    * Import nltk library
    * Import sent_tokenize to get sentence tokens.
    * Import word_tokenize to get word tokens.
    * Above two tokenizers can be imported from nltk.tokenize.
    
    Example:
    
    Sentence: Hi, Python is interesting. Learning python is fun.
    
    sent_tokenize(Sentence): list('Hi, Python is interesting.', 'Learning python is fun.')
    
    word_tokenize(Sentence): list('Hi', ',', 'Python', 'is', 'interesting', '.', 'Learning', 'python', 'is', 'fun', '.')
    
    

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# Sentence tokenization.
s_tokens = sent_tokenize(par)

# Word tokenization.
w_tokens = word_tokenize(par) 

# Display 5 sentence tokens.
for i in range(0,5): 
    print(s_tokens[i]) 

print("\n--------------------------------------------------------------------------------------------------") 

# Display 5 word tokens.
for i in range(0,5):
    print(w_tokens[i]) 
```

### 2. POS: Process of classifying the words in a text/corpus into their parts of speech and labeling them accordingly.
    
    Example: 
        
        The little yellow dog barked at the cat
    
        POS:
             The --> DT :: determiner
             little --> JJ :: adjective
             yellow --> JJ :: adjective
             dog --> NN :: noun
             barked --> VBD :: Verb
             at --> IN :: conjunction
             the --> DT :: determiner
             cat --> NN :: noun
         

```python
# Download tagsets and tagger.
nltk.download('tagsets')
nltk.download('averaged_perceptron_tagger')
```

```python
# Tag the word tokens with their respective POS. 
tagged = nltk.pos_tag(w_tokens)

# Display the tagged tokens.
print(tagged)
```

### 3. Stemming: Words are reduced to a root by removing the suffixes.
    
    Example:
        
        Applies --> Apply
        reading --> read
    

```python
# Download 'wordnet'.
nltk.download('wordnet')

# Import LancasterStemmer to perform stemming.
from nltk.stem import LancasterStemmer 

# Create an object for the Stemmer class to call stem method later.
stemmer = LancasterStemmer() 

s=[]

# Call the stem method and perform stemming on all the words present in w_tokens.
for i in w_tokens:
    s.append(stemmer.stem(i))

# Display the stemmed words.
print(s)
```

### 4. Lemmatization: Removes inflection by determining the part of speech using language database.
    
    Example:
        
        better --> good
        rocks -->rock
    

```python
# Import WordNetLemmatizer to perform Lemmatization.
from nltk.stem import WordNetLemmatizer

# Create an object for the Lemmatizer class to call lemmatize method later.
lemmatizer = WordNetLemmatizer()

l=[]

# Call the lemmatize method and perform Lemmatization on all the words present in w_tokens.
for i in w_tokens: 
    l.append(lemmatizer.lemmatize(i)) 

# Display the lemmatized words.
print(l)
 
```

### 5. Trigram: Set of co-occuring orcontinuous sequence of 3 items from the sequence of large text/sentence.
    
    Example:
        
        The quick red fox jumps over the lazy brown dog
        
        Trigram:
                 * The quick red
                 * quick red fox
                 * red fox jumps
                 * fox jumps over
                 * jumps over the
                 * over the lazy
                 * the lazy brown
                 * lazy brown dog
    
    

```python
# Import ngrams library
from nltk.util import ngrams 

# n value represents the type of ngrams. 3 for trigram, 2 for bigramm, .....
n = 3 
trigram = ngrams(par.split(),n) 

# Display the trigrams.
for i in trigram:
    print(i) 
```

### 6. Named Entity Recognition: Specificly recognizes the words into pre-defined categories.
    

```python
from nltk import wordpunct_tokenize, pos_tag, ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
# x is a tree that has all the tagged POS word tokens.
x = ne_chunk(pos_tag(wordpunct_tokenize(par))) 

# Dispaly NER applied words.
for i in x:
    print(i)
```

### 7. Plot the frequencies of words after removing the stop words.

Removing stop words: Frequent words that do not have specific semantics are removed.

```python
# Download stopwords set.
nltk.download('stopwords')
from nltk.corpus import stopwords  

# List all the words which aren't stop words.
w_tokens_without_sw = [word for word in w_tokens if not word in stopwords.words()] 

# Dispaly word tokens with out stop words.
print(w_tokens_without_sw)
```

```python
# Calculate the frequency of each word.
count = nltk.FreqDist(w_tokens_without_sw)

# Plot the graph between words and their frequency.
count.plot(len(w_tokens_without_sw),cumulative=False)
```

```python
# Plot 50 words with the frequency.
count.plot(50,cumulative=False)
```
